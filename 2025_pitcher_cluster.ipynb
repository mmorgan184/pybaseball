```python
# Import necessary libraries
import pandas as pd
from pybaseball import statcast_pitcher # Import the specific function we need

print("Libraries imported successfully.")
```

```python
  # Define the date range
start_date = '2025-04-01'
end_date = '2025-04-15'

print(f"Fetching Statcast data from {start_date} to {end_date}...")
```

```python
# Fetch the data
try:
    statcast_data = statcast(start_dt=start_date, end_dt=end_date, verbose=True)
    print("\nData fetched successfully!")
except Exception as e:
    print(f"\nAn error occurred during data fetching: {e}")
    statcast_data = pd.DataFrame() # Create an empty DataFrame on error
```

```python
# Check if data was loaded
if not statcast_data.empty:
    print("Performing basic DataFrame inspection...")

    # Display the first 5 rows
    print("\n--- First 5 Rows (head) ---")
    display(statcast_data.head())

    # Display the dimensions (rows, columns) of the DataFrame
    print(f"\n--- DataFrame Shape ---")
    print(f"Rows: {statcast_data.shape[0]}, Columns: {statcast_data.shape[1]}")

    # Display column names, non-null counts, and data types
    print("\n--- DataFrame Info ---")
    statcast_data.info()

else:
    print("\nNo data was loaded, skipping inspection.")
```
